```python
import numpy as np
import pandas as pd

def sum_by_D_and_T(a):
    df = pd.DataFrame(a)
    grouped = df.groupby(['D','M', 'T'])    
    sum = grouped.sum()
    return [([r[0], r[1],r[2], kv.to_dict().get('PM2MAX')]) for r, kv in sum.iterrows()]
```

```python
sum_by_D_and_T(a)
```

```python
b=sum_by_D_and_T(a)
```

```python
np.savetxt("Karvestatuesquare(1).txt",b,delimiter=',')
```
